##Feature mapping and generation

```python
df.drop(['ssc_b','hsc_b', 'hsc_s', 'degree_t'], axis=1, inplace=True)
```

```python
df["gender"] = df.gender.map({"M":0,"F":1})
df["workex"] = df.workex.map({"No":0, "Yes":1})
df["status"] = df.status.map({"Not Placed":0, "Placed":1})
df["specialisation"] = df.specialisation.map({"Mkt&HR":0, "Mkt&Fin":1})
```

```python
df.columns
```

##Additional feature generation

```python
def new_features(df):
    df['hsc_to_ssc'] = df['hsc_p'] / df['ssc_p']
    df['degree_to_hsc'] = df['degree_p'] / df['hsc_p']
    df['degree_to_ssc'] = df['degree_p'] / df['ssc_p']
    df['mba_to_degree'] = df['mba_p'] / df['degree_p']
    df['mba_to_etest'] = df['mba_p'] / df['etest_p']
    return df
```

```python
df = new_features(df)
```

##Model Prediction

## Train and Test Split

```python
# Seperating Features and Target
X = df.copy().drop('status', axis=1)
y = df['status']
```

```python
# scale each features
X_scaled = preprocessing.scale(X)
```

```python
#Train Test Split
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, y, test_size=0.3)
X_train.shape, Y_train.shape, X_test.shape
```

##KNN Classification

```python
# k-nearest neighbor
from sklearn.metrics import f1_score
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
knn_Y_pred = knn.predict(X_test)
knn_f1_score=f1_score(knn_Y_pred,Y_test)
knn_accuracy = knn.score(X_test, Y_test)
knn_accuracy
```

```python
knn_f1_score
```

##Confusion matrix

```python
# creating confusion matrix heatmap

conf_mat = confusion_matrix(Y_test, knn_Y_pred)
fig = plt.figure(figsize=(10,7))
group_names = ['True Neg','False Pos','False Neg','True Pos']
group_counts = ["{0:0.0f}".format(value) for value in
                conf_mat.flatten()]
labels = [f"{v1}\n{v2}" for v1, v2 in zip(group_names,group_counts)]
labels = np.asarray(labels).reshape(1,1)
sns.heatmap(conf_mat, annot=labels, annot_kws={"size": 16}, fmt='')
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
# getting precision, recall and f1-score via classification report

print(classification_report(Y_test, knn_Y_pred))
```

## Decision Tree Classification

```python
# Decision Tree
from sklearn.metrics import f1_score
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)
decision_tree_Y_pred = decision_tree.predict(X_test)
decision_tree_f1_score=f1_score(decision_tree_Y_pred,Y_test)
decision_tree_accuracy = decision_tree.score(X_test, Y_test)
decision_tree_accuracy
```
